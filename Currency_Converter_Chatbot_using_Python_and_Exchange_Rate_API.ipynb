```python
import requests

def convert_currency(amount,from_currency,to_currency):
  url = "https://api.frankfurter.app/latest"
  params ={
      "amount": amount,
      "from": from_currency,
      "to": to_currency
  }

  response = requests.get(url,params=params)
  data=response.json()


  result = data["rates"][to_currency]
  return f"{amount} {from_currency} = {result} {to_currency}"


print("Currency Chatbot")
print("Type 'exit' to stop\n")

while True:
  user_input = input("Ask me (e.g , convert 100 USD to NGN):")

  if user_input.lower()=="exit":
    break

  try:

    parts = user_input.split()
    amount = float(parts[1])
    from_currency = parts[2].upper()
    to_currency = parts[4].upper()


    answer = convert_currency(amount,from_currency,to_currency)
    print(answer)


  except:
    print("Try format: convert 100 USD to NGN")
```
